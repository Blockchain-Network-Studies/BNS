```python
#Read the csv file
df_AAVE=pd.read_csv('https://raw.githubusercontent.com/coinmetrics/data/master/csv/aave.csv')
df_AAVE['time']=pd.to_datetime(df_AAVE['time'])
df_AAVE.head()
```

```python
# define the time range for data analysis
import datetime
start = datetime.datetime(2020,10,9)
end = datetime.datetime(2023,7,30)
```

```python
# select data during the defined time range
df_AAVE = df_AAVE[(df_AAVE.time>=start) & (df_AAVE.time<=end)]
df_AAVE.head()
```

```python
# sort data by time
df_AAVE = df_AAVE.sort_values(by='time', ascending=True)
```

```python
# drop the replicated rows
df_AAVE =df_AAVE.dropna(axis=0,thresh=3)
df_AAVE.tail(10)
```

```python
df_eth = pd.read_csv('https://raw.githubusercontent.com/coinmetrics/data/master/csv/eth.csv')
df_eth['time']=pd.to_datetime(df_eth['time'])
df_eth.head()
```

```python
# define the time range for data analysis
import datetime
start = datetime.datetime(2020,10,9)
end = datetime.datetime(2023,7,30)
```

```python
# select data during the defined time range
df_eth = df_eth[(df_eth.time>=start) & (df_eth.time<=end)]
df_eth.head()
```

```python
# sort data by time
df_eth = df_eth.sort_values(by='time', ascending=True)
```

```python
# drop the replicated rows
df_eth =df_eth.dropna(axis=0,thresh=3)
df_eth.tail(10)
```

```python
df_eth = df_eth[['time','PriceUSD']]
df_eth = df_eth.rename(columns={'PriceUSD': 'ETHPriceUSD'})
```

```python
df_eth.head(10)
```

```python
Network_AAVE = pd.read_csv('AAVE_Network_Features.csv')
# convert time to date
Network_AAVE ['time']=pd.to_datetime(Network_AAVE ['time'])
# select time interval
Network_AAVE  = Network_AAVE [(Network_AAVE.time>start) &(Network_AAVE.time<=end)]
Network_AAVE = Network_AAVE .reset_index()
Network_AAVE = Network_AAVE .drop(['index','Unnamed: 0'], 1)
Network_AAVE['relative_degree'] = Network_AAVE['num_edges']/(Network_AAVE['num_nodes']*(Network_AAVE['num_nodes']-1)/2)
Network_AAVE.head()
```

```python
#df_AAVE =df_AAVE.merge(TVL_AAVE, left_on='time', right_on='date')
df_AAVE =df_AAVE.merge(Network_AAVE, left_on = 'time', right_on='time')
df_AAVE =df_AAVE.merge(df_eth, left_on = 'time', right_on='time')
df_AAVE.head()
```

```python
df_AAVE.to_csv("AAVE Merged Datasets.csv")
```
