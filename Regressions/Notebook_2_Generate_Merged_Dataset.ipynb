```python
#Read the csv file
df_AAVE=pd.read_csv('https://raw.githubusercontent.com/coinmetrics/data/master/csv/aave.csv')
df_AAVE['time']=pd.to_datetime(df_AAVE['time'])
df_AAVE.head()
```

```python
# define the time range for data analysis
import datetime
start = datetime.datetime(2020,10,9)
end = datetime.datetime(2021,10,10)
```

```python
# select data during the defined time range
df_AAVE = df_AAVE[(df_AAVE.time>=start) & (df_AAVE.time<=end)]
df_AAVE.head()
```

```python
# sort data by time
df_AAVE = df_AAVE.sort_values(by='time', ascending=True)
df_AAVE.tail(10)
```

```python
# drop the replicated rows
df_AAVE =df_AAVE.dropna(axis=0,thresh=3)
df_AAVE.tail(10)
```

```python
import pandas as pd
import requests
import json 
from datetime import datetime
TVL_AAVE = requests.get('https://data-api.defipulse.com/api/v1/defipulse/api/GetHistory?resolution=history&project=aave&api-key=5eafa87608af1f342443fc1fc293d5a596f3aa509526c9c0f663a1b5723a')
TVL_AAVE = TVL_AAVE.json()
TVL_AAVE =pd.DataFrame(TVL_AAVE)
TVL_AAVE['project']='AAVE'
TVL_AAVE.head()
```

```python
Network_AAVE = pd.read_csv('/content/drive/MyDrive/Network Study/Data/AAVE_Network_Features.csv')
# convert time to date
Network_AAVE ['time']=pd.to_datetime(Network_AAVE ['time'])
# select time interval
Network_AAVE  = Network_AAVE [(Network_AAVE.time>start) &(Network_AAVE.time<=end)]
Network_AAVE = Network_AAVE .reset_index()
Network_AAVE = Network_AAVE .drop(['index','Unnamed: 0'], 1)
Network_AAVE['relative_degree'] = Network_AAVE['num_edges']/(Network_AAVE['num_nodes']*(Network_AAVE['num_nodes']-1)/2)
Network_AAVE.head()
```

```python
#Epoch & Unix Timestamp Conversion Tools
import datetime
TVL_AAVE['timestamp']=TVL_AAVE['timestamp'].astype(int)
TVL_AAVE['time']=TVL_AAVE['timestamp'].apply(lambda x: datetime.datetime.utcfromtimestamp(x).replace(tzinfo=datetime.timezone.utc))

TVL_AAVE['date'] = pd.to_datetime(TVL_AAVE['time']).dt.date
TVL_AAVE=TVL_AAVE.drop(columns=['time','timestamp'])
TVL_AAVE['date']=pd.to_datetime(TVL_AAVE['date'])

TVL_AAVE.head()
```

```python
# select time interval of aave_TVL that is consistent to the other two datasets
TVL_AAVE = TVL_AAVE[(TVL_AAVE.date>start) &(TVL_AAVE.date<=end)]
TVL_AAVE = TVL_AAVE.sort_values(by='date', ascending=True)
TVL_AAVE.head()
```

```python
df_AAVE =df_AAVE.merge(TVL_AAVE, left_on='time', right_on='date')
df_AAVE =df_AAVE.merge(Network_AAVE, left_on = 'time', right_on='time')
df_AAVE.head()
```

```python
df_AAVE.to_csv("AAVE Merged Datasets.csv")
```
